```python
# 고양이, 강아지, 기니피그 이미지를 크롤링으로 저장하기.

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import urllib.request
import time
import urllib.request


# 고양이, 강아지, 기니피그 이미지를 각각 100장 저장.

animal = ['고양이', '강아지', '기니피그']

driver = webdriver.Chrome()
driver.get("https://www.google.co.kr/imghp?hl=ko&ogbl")
elem = driver.find_element("name", "q")
elem.clear()
elem.send_keys(animal[0])
elem.submit()


last_height = driver.execute_script("return document.body.scrollHeight")

while True:

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    time.sleep(3)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_element(By.CSS_SELECTOR, ".mye4qd").click()
        except:
            print("표시할 결과가 없습니다!")
            break
    last_height = new_height

# 고양이 이미지 저장하기

print('이미지 저장을 실행하겠습니다.')

images = driver.find_elements(By.CSS_SELECTOR, "a.wXeWr.islib.nfEiy")
count = 1

for image in images:
    image.send_keys(Keys.RETURN)
    time.sleep(2)
    imgUrl = driver.find_element(
        By.CSS_SELECTOR, ".n3VNCb").get_attribute("src")
    urllib.request.urlretrieve(
        imgUrl, "./02.crawling_img/" + animal[0] + str(count) + ".jpg")
    try:
        print(str(count)+'번 ' + f'{animal[0]} 이미지를 저장하였습니다.')
    except:
        print(str(count)+'번 이미지 다운 중 문제 발생하였습니다.')
    count = count + 1
    if count > 100:
        print(animal[0]+"의 이미지를 모두 저장하였습니다.")
        break

# 강아지 이미지 저장하기
elem = driver.find_element("name", "q")
elem.clear()
elem.send_keys(animal[1])
elem.submit()


last_height = driver.execute_script("return document.body.scrollHeight")

while True:

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    time.sleep(3)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_element(By.CSS_SELECTOR, ".mye4qd").click()
        except:
            print("표시할 결과가 없습니다!")
            break
    last_height = new_height

print('이미지 저장을 실행하겠습니다.')

images = driver.find_elements(By.CSS_SELECTOR, "a.wXeWr.islib.nfEiy")
for image in images:
    image.send_keys(Keys.RETURN)
    time.sleep(2)
    imgUrl = driver.find_element(
        By.CSS_SELECTOR, ".n3VNCb").get_attribute("src")
    urllib.request.urlretrieve(
        imgUrl, "./02.crawling_img/" + animal[1] + str(count) + ".jpg")
    try:
        print(str(count)+'번 ' + f'{animal[1]} 이미지를 저장하였습니다.')
    except:
        print(str(count)+'번 이미지 다운 중 문제 발생하였습니다.')
    count = count + 1
    if count > 200:
        print(animal[1]+"의 이미지를 모두 저장하였습니다.")
        break



# 기니피그 이미지 저장하기
elem = driver.find_element("name", "q")
elem.clear()
elem.send_keys(animal[2])
elem.submit()


last_height = driver.execute_script("return document.body.scrollHeight")

while True:

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    time.sleep(3)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_element(By.CSS_SELECTOR, ".mye4qd").click()
        except:
            print("표시할 결과가 없습니다!")
            break
    last_height = new_height

print('이미지 저장을 실행하겠습니다.')

images = driver.find_elements(By.CSS_SELECTOR, "a.wXeWr.islib.nfEiy")

for image in images:
    image.send_keys(Keys.RETURN)
    time.sleep(2)
    imgUrl = driver.find_element(
        By.CSS_SELECTOR, ".n3VNCb").get_attribute("src")
    urllib.request.urlretrieve(
        imgUrl, "./02.crawling_img/" + animal[2] + str(count) + ".jpg")
    try:
        print(str(count)+'번 ' + f'{animal[2]} 이미지를 저장하였습니다.')
    except:
        print(str(count)+'번 이미지 다운 중 문제 발생하였습니다.')
    count = count + 1
    if count > 300:
        print(animal[2]+"의 이미지를 모두 저장하였습니다.")
        break


driver.close()
```
